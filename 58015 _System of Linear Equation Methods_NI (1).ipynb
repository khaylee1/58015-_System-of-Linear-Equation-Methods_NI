## Problem
The herds are split up as follows:<br>
At Mamie‚Äôs pasture: 16 cows, 13 heifers, 7 calves<br>
At McLoed‚Äôs pasture: 18 cows 11 calves 1 bull<br>
At Mary Lou‚Äôs pasture: 16 cows, 5 calves, 1 bull<br>
At Nadine‚Äôs pasture: 12 calves<br>
<br>
Cows make up what percentage of all the animals in the four pastures?<br>
Heifers make up what percentage of all the animals in the four pastures?<br>
Calves make up what percentage of all the animals in the four pastures?<br>
Bulls make up what percentage of all the animals in the four pastures?<br>

### Standard bracketed form
$Let : \\ C = Cow,  H= Heifers, c = Calves, B= Bulls\\\\
Mamie: \{16C + 13H + 7c + 0B = 36 \} \\
McLoed:\{ 18C + 0H + 11c + 1B = 30\} \\
Mary Lou: \{16C + 0H + 5c+ 1B = 22 \}\\
Nadine: \{0C + 0H + 12c + 0B =12\}$

### Linear Combination form (Matrix)

 $$ùëÄùëéùëöùëñùëí= \begin{bmatrix}16\\13\\7\end{bmatrix}, ùëÄùëêùêøùëúùëíùëë= \begin{bmatrix} 18\\11\\1\end{bmatrix},ùëÄùëéùëüùë¶ùêøùëúùë¢ = \begin{bmatrix}16\\5\\1\end{bmatrix},ùëÅùëéùëëùëñùëõùëí= \begin{bmatrix}12\end{bmatrix}$$


### Vector Visualization




```python
fig_1=plt.figure(1,figsize=(20,10.6))

vecta = np.array([16,13,7,0])
vectb = np.array([18,0,11,1])
vectc= np.array([16,0,5,1])
vectd= np.array([0,0,12,0])
total_animal=np.array([36,30,22,12]) # total animal in each pasture

plt.plot(total_animal,vecta) #mamie
plt.plot(total_animal,vectb) #mcloed
plt.plot(total_animal,vectc) #mary lou
plt.plot(total_animal,vectd) # nadine
plt.title("herds split up")
plt.xlabel("Total Number of Herds in Each Pasture")
plt.ylabel("Number of Herds in Each Pasture")
plt.legend(["mamie","mcloed","mary lou","nadine"])
plt.show()

```

```python
pastures =np.array([
    [16,13,7,0],
    [18,0,11,1],
    [16,0,5,1],
    [0,0,12,0]
])
total_animals=np.array([
    [36],
    [30],
    [22],
    [12]
    
])

animals = np.linalg.inv(pastures) @ total_animals
print('The percentage of cows in 4 pasture is:{:.2f}%'.
      format(float(animals[0]*50))) # cow total= 50
print('The percent of heifers in 4 pasture is: {:.2f}%'.
      format(float(animals[1]*13)))# heifers total = 13
print('The percent of calves in 4 pasture is:{:.2f}%'.
      format(float(animals[2]*35)))# calves total= 35
print('The percent of bulls in 4 pasture is:{:.2f}%'.
      format(float(animals[3]*2)))  # bulls total= 2
```
